```python
import requests
import json
```

```python
# Load and view the structure of the JSON file
with open("../data/data.json", "r") as file:
    data = json.load(file)
```

```python
def print_json_structure(data, indent=0):
    if isinstance(data, dict):
        for key, value in data.items():
            print(" " * indent + str(key))
            print_json_structure(value, indent + 2)
    elif isinstance(data, list):
        for i, item in enumerate(data):
            print(" " * indent + f"[{i}]")
            print_json_structure(item, indent + 2)
    else:
        print(" " * indent + str(data))


# Function to print the first few records
def print_first_records(data, num_records=3, indent=0):
    if isinstance(data, list):
        for i, item in enumerate(data[:num_records]):
            print(f"Record {i + 1}:")
            print_json_structure(item, indent + 2)
            print("\n")
    else:
        print_json_structure(data)
```
